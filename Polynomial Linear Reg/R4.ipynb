```r
dataset = read.csv('Position_Salaries.csv')
head(dataset)
dataset = dataset[2:3]
```

```r
#fitting linear reg
lreg = lm(formula = Salary ~., data = dataset)
```

```r
#fitting polynomial reg of degree 4
dataset$Level2 = dataset$Level^2
dataset$Level3 = dataset$Level^3
dataset$Level4 = dataset$Level^4
polyreg = lm(formula = Salary~., data=dataset)
```

```r
#visualising linear reg
library(ggplot2)
ggplot()+
    geom_point(aes(x= dataset$Level, y= dataset$Salary), color= 'red')+
    geom_line(aes(x= dataset$Level, y= predict(lreg, newdata= dataset)), color= 'blue')+
    ggtitle('Linear Regression')+
    xlab('Level')+
    ylab('Salary')
```

```r
#visualising Polynomial reg with smooother curve
library(ggplot2)
x_grid = seq(min(dataset$Level),max(dataset$Level),0.1)
ggplot()+
    geom_point(aes(x= dataset$Level, y= dataset$Salary), color= 'red')+
    geom_line(aes(x= x_grid, y= predict(polyreg, newdata= data.frame(Level = x_grid, 
                                                                     Level2= x_grid^2, 
                                                                     Level3= x_grid^3,
                                                                     Level4= x_grid^4))), color= 'blue')+
    ggtitle('Polynoial Regression')+
    xlab('Level')+
    ylab('Salary')
```

### Predicting new results

```r
#with linear regression
predict(lreg, newdata = data.frame(Level =6.5))
```

```r
#with polynomial regression
predict(polyreg, newdata = data.frame(Level=6.5,
                                     Level2=6.5^2,
                                     Level3=6.5^3,
                                     Level4=6.5^4))
```
