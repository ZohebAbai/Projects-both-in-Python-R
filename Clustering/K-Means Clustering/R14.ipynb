```r
# Importing the dataset
dataset = read.csv('Mall_Customers.csv')
head(dataset)
dataset = dataset[4:5]
```

```r
#Using elbow method to find optimal numbers of clusters
set.seed(101)
wcss <- vector()
for (i in 1:10) wcss[i] <- sum(kmeans(dataset, i)$withinss)
plot( 1:10, wcss, type='b', main= paste('The Elbow Method'), xlab= 'No of Clusters', ylab= 'WCSS' )
```

```r
# Based on above graph we consider number of cluster as 6
# Fitting K-Means to the dataset
set.seed(101)
kmeans= kmeans(x= dataset, centers=6)
y_kmeans = kmeans$cluster
```

```r
#visualizing the cluster
library(cluster)
clusplot(dataset, y_kmeans, lines=0, shade=TRUE, color= TRUE, labels=4, plotchar= FALSE, span= TRUE, 
          main= paste('Clusters of Customers'), xlab= 'Annual Income', ylab= 'Spending Score' )
```
