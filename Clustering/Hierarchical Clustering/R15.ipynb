```r
# Importing the dataset
dataset = read.csv('Mall_Customers.csv')
dataset = dataset[4:5]
head(dataset)
#We are going to cluster our customers based on Annual Income and Spending Scores
```

#### We shall consider Agglomerative (Bottom Up) Approach for Hierarchical Clustering. 

```r
# Using the dendrogram to find the optimal number of clusters. Ward is for minimizing variance within each cluster
dendrogram = hclust(d= dist(dataset, method = 'euclidean'), method = 'ward.D') 
plot(dendrogram, main = paste('Dendrogram'), xlab = 'Customers', ylab = 'Euclidean distances')
```

#### For Threshold in dendogram we consider the longest vertical line that doesn't cross any extended horizontal line.
#### Here following this process, it gives us 6 clusters.

```r
# Fitting Hierarchical Clustering to the dataset
hc = hclust(d = dist(dataset, method = 'euclidean'), method = 'ward.D')
y_hc = cutree(hc, 6)
```

```r
# Visualising the clusters
library(cluster)
clusplot(dataset,
         y_hc,
         lines = 0,
         shade = TRUE,
         color = TRUE,
         labels= 4,
         plotchar = FALSE,
         span = TRUE,
         main = paste('Clusters of customers'),
         xlab = 'Annual Income',
         ylab = 'Spending Score')
```

#### Here Cluster no. 6 is our target customer
