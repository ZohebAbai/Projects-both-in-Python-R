```r
df = read.csv('Position_Salaries.csv')
df = df[2:3]
head(df)
```

```r
library(rpart)
reg = rpart(formula = Salary~., data= df, control = rpart.control(minsplit = 1))
```

```r
y_pred = predict(reg, data.frame(Level= 6.5))
y_pred
```

```r
# Visualising the Decision Tree Regression results (higher resolution)
library(ggplot2)
x_grid = seq(min(df$Level),max(df$Level),0.01)
ggplot()+
    geom_point(aes(x= df$Level, y= df$Salary),color= 'red')+
    geom_line(aes(x= x_grid, y=predict(reg, data.frame(Level = x_grid))), color= 'blue')+
    ggtitle('Decision Tree Regressor')+
    xlab('Position Level')+
    ylab('Salary')
```

```r
# Plotting the tree
plot(reg)
text(reg)
```
