```r
df = read.csv('Position_Salaries.csv')
df =df[2:3]
head(df)
```

```r
library(randomForest)
set.seed(101)
reg = randomForest(x = df[1], y= df$Salary, ntree= 500)
```

```r
y_pred = predict(reg, data.frame(Level= 6.5))
y_pred
```

```r
# Visualising the Random Forest Regression results with higher resolution
library(ggplot2)
x_grid = seq(min(df$Level), max(df$Level), 0.01)
ggplot() +
  geom_point(aes(x = df$Level, y = df$Salary),colour = 'red') +
  geom_line(aes(x = x_grid, y = predict(reg, data.frame(Level = x_grid))),colour = 'blue') +
  ggtitle('Random Forest Regression') +
  xlab('Level') +
  ylab('Salary')
```
